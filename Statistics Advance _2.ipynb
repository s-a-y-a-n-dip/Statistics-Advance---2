```python
#1. What is hypothesis testing in statistics?
#ANS.
#Hypothesis testing is a statistical method used to make decisions or inferences about population parameters based on sample data.
#It involves testing an assumption (hypothesis) to determine whether there's enough evidence to support it.
```

```python
#2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
#ANS.
#Null Hypothesis (H₀): Assumes no effect or no difference.
#Alternative Hypothesis (H₁): Assumes there is an effect or difference.
#Hypothesis testing evaluates whether sample data provides sufficient evidence to reject the null hypothesis in favor of the alternative.
```

```python
#3. What is the significance level in hypothesis testing, and why is it important?
#The significance level (α) is the probability of rejecting the null hypothesis when it is true (Type I error). 
#Common values are 0.05 or 0.01. It defines the threshold for statistical significance.
```

```python
#4. What does a P-value represent in hypothesis testing?
#ANS.
#The P-value is the probability of obtaining results at least as extreme as those observed, assuming the null hypothesis is true.
#It quantifies the evidence against the null hypothesis.
```

```python
#5. How do you interpret the P-value in hypothesis testing?
#ANS.
#Low P-value (≤ α): Strong evidence against the null → Reject H₀
#High P-value (> α): Weak evidence against the null → Fail to reject H₀
#It helps in deciding whether to accept or reject the null hypothesis.
```

```python
#6. What are Type I and Type II errors in hypothesis testing?
#ANS.
#Type I Error (False Positive): Rejecting H₀ when it is true
#Type II Error (False Negative): Failing to reject H₀ when it is false
#Minimizing both is crucial for reliable testing.
```

```python
#7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
#ANS.
#One-tailed test: Tests for effect in one direction (e.g., greater than or less than)
#Two-tailed test: Tests for effect in both directions (e.g., not equal to)
#The choice depends on the research question or hypothesis.
```

```python
#1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

from scipy import stats
import numpy as np

sample = np.array([105, 110, 98, 115, 102])
population_mean = 100
population_std = 10

z_stat = (np.mean(sample) - population_mean) / (population_std / np.sqrt(len(sample)))
p_value = 1 - stats.norm.cdf(z_stat)

print(f"Z-statistic: {z_stat:.2f}, P-value: {p_value:.4f}")
```

```python
#2. Simulate random data to perform hypothesis testing and calculate the corresponding P-value using Python.

np.random.seed(0)
data = np.random.normal(loc=50, scale=5, size=100)
t_stat, p_val = stats.ttest_1samp(data, 52)
print(f"T-statistic: {t_stat:.2f}, P-value: {p_val:.4f}")
```

```python
#3.  Implement a one-sample Z-test using Python to compare the sample mean with the population mean.

def one_sample_z_test(sample, pop_mean, pop_std):
    z = (np.mean(sample) - pop_mean) / (pop_std / np.sqrt(len(sample)))
    p = 2 * (1 - stats.norm.cdf(abs(z)))
    return z, p

sample_data = np.random.normal(100, 15, 30)
z_score, p_val = one_sample_z_test(sample_data, 100, 15)
print(f"Z-score: {z_score:.2f}, P-value: {p_val:.4f}")
```

```python
#4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

import matplotlib.pyplot as plt

z = 2.0  # example Z-score
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x)

plt.plot(x, y)
plt.fill_between(x, y, where=(x < -z) | (x > z), color='red', alpha=0.5)
plt.title("Two-tailed Z-test Region")
plt.show()
```

```python
#5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

def visualize_errors():
    x = np.linspace(-4, 4, 1000)
    y1 = stats.norm.pdf(x, 0, 1)     # Null hypothesis distribution
    y2 = stats.norm.pdf(x, 1.5, 1)   # Alternative hypothesis

    plt.plot(x, y1, label='H0')
    plt.plot(x, y2, label='H1')
    plt.fill_between(x, y1, where=(x > 1.64), color='red', alpha=0.3, label='Type I Error')
    plt.fill_between(x, y2, where=(x <= 1.64), color='blue', alpha=0.3, label='Type II Error')
    plt.legend()
    plt.title("Type I and Type II Errors")
    plt.show()

visualize_errors()
```

```python
#9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    stderr = stats.sem(data)
    h = stderr * stats.t.ppf((1 + confidence) / 2, n - 1)
    return mean - h, mean + h

data = np.random.normal(100, 15, 30)
ci = confidence_interval(data)
print(f"95% Confidence Interval: {ci}")
```

```python
#10.Write a Python program to calculate the margin of error for a given confidence level using sample data.

import numpy as np
import scipy.stats as stats

def margin_of_error(data, confidence_level=0.95):
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)  # Standard Error of the Mean
    z_score = stats.norm.ppf((1 + confidence_level) / 2)
    margin = z_score * std_err
    return margin

# Example usage:
sample_data = np.random.normal(50, 10, 100)  # Sample data
print(f"Margin of Error: {margin_of_error(sample_data):.4f}")
```

```python
#11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

def bayes_theorem(prior_A, prob_B_given_A, prob_B_given_not_A):
    # P(A|B) = P(B|A) * P(A) / P(B)
    prob_B = (prob_B_given_A * prior_A) + (prob_B_given_not_A * (1 - prior_A))
    posterior_A_given_B = (prob_B_given_A * prior_A) / prob_B
    return posterior_A_given_B

# Example: Probability of disease given a positive test
prior = 0.01  # P(Disease)
prob_pos_given_disease = 0.95  # P(Test+|Disease)
prob_pos_given_no_disease = 0.05  # P(Test+|No Disease)
posterior = bayes_theorem(prior, prob_pos_given_disease, prob_pos_given_no_disease)
print(f"Posterior Probability: {posterior:.4f}")
```

```python
#22. Perform a one-way ANOVA test in Python and visualize the results with boxplots to compare group means.

import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

# Perform one-way ANOVA
def one_way_anova(data, group_col, value_col):
    groups = data[group_col].unique()
    group_data = [data[data[group_col]==g][value_col] for g in groups]
    f_val, p_val = stats.f_oneway(*group_data)
    print(f"F-value: {f_val}, p-value: {p_val}")
    return f_val, p_val

# Visualize the comparison
def visualize_groups(data, group_col, value_col, title="Group Name Comparison"):
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=data, x=group_col, y=value_col)
    plt.title(title)
    plt.show()

# Example usage:
# Assuming 'data' is your DataFrame with 'Group' and 'Value' columns
# f_val, p_val = one_way_anova(data, "Group", "Value")
# visualize_groups(data, "Group", "Value")
```
